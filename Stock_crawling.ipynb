```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
# 182~236

def main(code_number):
    BASE_URL='https://finance.naver.com/sise/sise_group_detail.nhn?'
    START_PAGE = code_number
    END_PAGE = 236
    fields = []
    j = 1
    res = requests.get(BASE_URL + "&type=" + "upjong" + "&no=" + str(START_PAGE))
    page_soup = BeautifulSoup(res.text, 'lxml')
    ipt_html = page_soup.select_one('div.subcnt_sise_item_top')
    fields = [item.get('value') for item in ipt_html.select('input')]


    data = {'menu': 'upjong',
            'fieldIds':  fields,
            'returnUrl': BASE_URL + "&type=" + "upjong" + "&no=" + str(START_PAGE)}

    res = requests.post('https://finance.naver.com/sise/field_submit.nhn', data=data)
    page_soup = BeautifulSoup(res.text, 'lxml')

    upjong = page_soup.select('div.bnd_wid > table > tr > td')
    table_html = page_soup.select('div.box_type_l > table > tbody > tr > td.number')
    table_html1 = page_soup.select('div.box_type_l > table > thead > tr > th')
    name = page_soup.select('div.box_type_l > table > tbody > tr > td.name')
    name1 = page_soup.select('div.box_type_l > table > tbody > tr > td')
    catagory = upjong[1].text.strip()

    result_name = []
    length = len(name)
    for i in range(0,length):
        a =  name[i].text.strip()
        result_name.append(a)

    result_fields = []
    text = ""
    for i in range(0,31):
        a =  table_html1[i].text.strip()
        result_fields.append(a)

    result_value = []
    text = ""
    for i in range(0,30*length):
        a =  table_html[i].text.strip()
        result_value.append(a)

    result = []
    for i in range(0,len(name)*32):
        a = name1[i].text.strip()
        result.append(a)
        
    result_fields[result_fields.index('거래대금')] = '거래대금(백만)'
    result_fields[result_fields.index('상장주식수')] = '상장주식수(천주)'
    result_fields[result_fields.index('시가총액')] = '시가총액(억)'
    result_fields[result_fields.index('자산총계')] = '자산총계(억)'
    result_fields[result_fields.index('부채총계')] = '부채총계(억)'
    result_fields[result_fields.index('매출액')] = '매출액(억)'
    result_fields[result_fields.index('영업이익')] = ' 영업이익(억)'
    result_fields[result_fields.index('당기순이익')] = '당기순이익(억)'
    result_fields[result_fields.index('주당순이익')] = '주당순이익(원)'
    result_fields[result_fields.index('보통주배당금')] = '보통주배당금(원)'
    result_fields[result_fields.index('PER')] = 'PER(배)'
    result_fields[result_fields.index('ROE')] = 'ROE(%)'
    result_fields[result_fields.index('ROA')] = 'ROA(%)'
    result_fields[result_fields.index('PBR')] = 'PBR(배)'
    result_fields[result_fields.index('유보율')] = '유보율(%)'

    data = np.array(result)
    data1 = data.reshape(length,32)
    data1 = np.delete(data1,(31), axis=1)
    vars()['df{}'.format(j)] = data1
    vars()['df{}'.format(j)] = pd.DataFrame(data=data1, columns=result_fields)
    vars()['df{}'.format(j)].to_excel("stock.xlsx", sheet_name = catagory, header = True, startrow = 1)
    return vars()['df{}'.format(j)]
```

```python
main(202)
```
